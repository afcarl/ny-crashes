```python
%matplotlib inline
import matplotlib as plt
import pandas as pd
```

```python
cases = pd.read_csv('Motor_Vehicle_Crashes_-_Case_Information__Three_Year_Window.csv')
individuals = pd.read_csv('Motor_Vehicle_Crashes_-_Individual_Information__Three_Year_Window.csv',
                          index_col='Case Individual ID')
vehicles = pd.read_csv('Motor_Vehicle_Crashes_-_Vehicle_Information__Three_Year_Window.csv',
                       index_col='Case Vehicle ID')
violations = pd.read_csv('Motor_Vehicle_Crashes_-_Violation_Information__Three_Year_Window.csv')
```

```python
cases = cases.set_index(pd.to_datetime(cases['Date'] + ' ' + cases['Time'].fillna('')))
```

```python
cases['Year'].resample('M').count().plot()
```

```python
cases['Crash Descriptor'].value_counts().plot(kind='bar')
```

```python
cases['Year'].value_counts().sort_index().plot(kind='bar')
```

```python
cases[cases['Crash Descriptor'] == 'Fatal Accident']['Year'].value_counts().sort_index().plot(kind='bar')
```

```python
vehicles['Vehicle Body Type'].value_counts()[:10].plot(kind='bar')
```

```python
crashes = individuals.merge(vehicles, left_on='Case Vehicle ID', right_index=True)
```

```python
crashes[crashes['Injury Severity'] == 'Killed']['Vehicle Body Type'].value_counts()[:10].plot(kind='bar')
```

```python
bikes = crashes[crashes['Vehicle Body Type'] == 'BICYCLIST']
motorcycles = crashes[crashes['Vehicle Body Type'] == 'MOTORCYCLE']
```

```python
bikes['Injury Severity'].value_counts().plot(kind='bar')
```

```python
motorcycles['Injury Severity'].value_counts().plot(kind='bar')
```
